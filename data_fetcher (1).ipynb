```python
import os
import requests
import pandas as pd
from tqdm import tqdm


MAPBOX_TOKEN = "........oYW5pMjMxMSn......." #my token
IMG_SIZE = 224        # MUST be int
ZOOM = 18
OUT_DIR = "images/train"
#images/test for test dataset
CSV_PATH = "train.csv  
#test2.csv for test dataset

# ---------------------------------------

os.makedirs(OUT_DIR, exist_ok=True)

df = pd.read_csv(CSV_PATH)


def fetch_image(lat, lon, save_path):
    if os.path.exists(save_path):
        return

    url = (
        f"https://api.mapbox.com/styles/v1/mapbox/satellite-v9/static/"
        f"{lon},{lat},{ZOOM}/{IMG_SIZE}x{IMG_SIZE}"
        f"?access_token={MAPBOX_TOKEN}"
    )

    try:
        r = requests.get(url, timeout=10)

        if r.status_code == 200 and "image" in r.headers.get("Content-Type", ""):
            with open(save_path, "wb") as f:
                f.write(r.content)
        else:
            print(f"Failed ({r.status_code}) for {lat}, {lon}")

    except Exception as e:
        print(f"Error for {lat}, {lon}: {e}")


def download_dataset(df):
    for idx, row in tqdm(df.iterrows(), total=len(df)):
        save_path = os.path.join(OUT_DIR, f"{idx}.png")
        fetch_image(row["lat"], row["long"], save_path)


download_dataset(df)
```
